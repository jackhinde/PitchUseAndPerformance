# User-Defined Functions

```python
def create_binary_pitch_type_column_from_index(df, pitch_type):
    '''
    Creates a column in 'df' with column name is_'pitch_type' having value 1 if the pitch type is of type 'pitch_type' and value 0 if the pitch type is not of type 'pitch_type'.
    Searches on a MultiIndex where the pitch type is the fourth level of the index.
    
            Parameters:
                    df (MultiIndex): A MultiIndex containing the pitch type in the third level of the index.
                    pitch_type (str): The pitch type to be searched for as a string. The name of the created column will be is_'pitch_type'.
    '''
    # create column name by concatenating is_ and pitch_type
    column_name = 'is_' + pitch_type
    # add boolean column if it is of type pitch_type
    # set level=3 to search in 'pitch_type' index
    df[column_name] = df.index.isin([pitch_type], level=3)
    # convert to binary
    df[column_name] = df[column_name].astype(int)
```

```python
def create_binary_pitch_use_column(df, pitch_use):
    '''
    Creates a column in 'df' with column name is_'pitch_use' having value 1 if the pitch type has usage rank 'pitch_use' and value 0 if the pitch type is does not have usage rank 'pitch_use'.
    
            Parameters:
                    df (DataFrame): A DataFrame containing a column called 'use' containing the pitch type's usage rank.
                    pitch_use (str): The usage rank to be searched for as a string. The name of the created column will be is_'pitch_use'.
    '''
    # create column name by concatenating is_ and pitch_use
    column_name = 'is_' + pitch_use
    # add boolean column for pitch use
    df[column_name] = df['use'] == pitch_use
    # convert to binary
    df[column_name] = df[column_name].astype(int)
```

```python
def use_pybaseball_playerid_lookup(first_name, last_name):
    '''
    Returns a DataFrame containing the first name, last name, MLBAM ID, debut year, and last year played of the player(s) best matching the first and last name of the search using pybaseball's playerid_lookup function.
    This is a simple wrapper around the playerid_lookup function from pybaseball that formats results of the search for players in the 2021-2023 seasons.
    If an exact match is found the function will only return one player.
    This function only returns players who appeared between 2021 and 2023. In the case that no player is found None will be returned.
    More information on playerid_lookup can be found at https://github.com/jldbc/pybaseball.
    
        Parameters:
                first_name (str): The first name of the player to be searched for.
                last_name (str): The last name of the player to be searched for.
                
        Returns:
                player_ids (DataFrame): A DataFrame containing the first name, last name, MLBAM ID, debut year and last year played of all players identified in the search.
                None: Returned if no player who appeared between 2021 and 2023 was found.
    '''
    # use pybaseball's playerid_lookup
    # set fuzzy=True to default to expecting a on-exact search result
    # if an exact match is found then only one player will be returned
    player_ids = playerid_lookup(last_name, first_name, fuzzy=True)
    # change the name of the MLBAM ID column for easier interpretation
    player_ids = player_ids.rename(columns={'key_mlbam': 'mlbam_id'})
    # filter to remove players without a finalyear played
    # this is mostly players who played long before 2021
    # current players will have mlb_played_last equal to the current year
    player_ids = player_ids[player_ids['mlb_played_last'] != '']
    # convert years to int
    player_ids['mlb_played_first'] = player_ids['mlb_played_first'].astype(int)
    player_ids['mlb_played_last'] = player_ids['mlb_played_last'].astype(int)
    # if there are multiple players found in the search (a non-exact match was needed)
    if (len(player_ids.index) > 1):
        # print the number of those players who appeared between 2021 and 2023
        print(f"Of these 5 most similar names, {len(player_ids.loc[((player_ids['mlb_played_last'] >= 2021) & (player_ids['mlb_played_first'] <= 2023)),:].index)} played between 2021 and 2023.")
        # if none of those players appeared between 2021 and 2023 return None
        if (len(player_ids.loc[((player_ids['mlb_played_last'] >= 2021) & (player_ids['mlb_played_first'] <= 2023)),:].index) == 0):
            return None
    # filter to keep only players who appeared between 2021 and 2023
    player_ids = player_ids.loc[((player_ids['mlb_played_last'] >= 2021) & (player_ids['mlb_played_first'] <= 2023)), ['name_first', 'name_last', 'mlbam_id', 'mlb_played_first', 'mlb_played_last']]
    # if there are no found players who appeared between 2021 and 2023 return None
    # this is already tested for for a non-exact search result, but there was no test for this in an exact search result
    if (len(player_ids.index) == 0):
        print("An identicallly matched name found! They did not play between 2021 and 2023.")
        return None
    # return the player(s) matching the search result who appeared between 2021 and 2023
    return(player_ids)
```

```python
def view_pred_RV_per_100_diff(season, pitcher_id, calculation_type='constant_increase', constant_increase_val=0.05, 
                              nonconstant_pt_increase_list=None, pt_target_mix_list=None):
    '''
    Returns a DataFrame containing the percentage thrown, run value per 100 pitches, and predicted change to run value per 100 pitches based on the change to the pitch type's percentage thrown of the pitcher with MLBAM ID 'pitcher_id' pitch types in 'season'.
    The increase or decrease in percentage thrown of each of their pitch types must be specified and four different shortcuts for specified levels of change are supported.
    The four shortcuts for calculations are:
        'constant_increase' (default): Increasing or decreasing the use of each of a pitcher's pitch types by the same amount.
        'nonconstant_pt_increase': Increasing or decreasing the use of each of a pitcher's pitch types by different amounts.
        'pt_equal_mix': Throwing each of a pitcher's pitch types the same amount of time.
        'pt_target_mix': Throwing each pitch type some proportion of time that is not the same for each of their pitch types.
    Examples of questions that may be answered by each of the shortcuts are:
        'constant_increase': How would each of this pitcher's pitch types be expected to perform if they were thrown 5 percentage points more often?
        'nonconstant_pt_increase': How would each of this pitcher's pitch types be expected to perform if they threw their changeup 5 percentage points more often, their fastball 8 percentage points less often, and their slider 3 percentage points more often?
        'pt_equal_mix': How would each of this pitcher's pitch types be expected to perform if they threw all of their pitch types equally?
        'pt_target_mix': How would each of this pitcher's pitch types be expected to perform if they threw their changeup 20% of the time, their fastball 45% of the time and their slider 35% of the time?
    
        Parameters:
                season (int): One of 2021, 2022 or 2023.
                pitcher_id (int): The MLBAM ID of the pitcher.
                calculation_type (str): One of 'constant_increase', 'nonconstant_pt_increase', 'pt_equal_mix', 'pt_target_mix'. Defaults to 'constant_increase'.
                constant_increase_val (float): The increase or decrease in percentage thrown to be applied to each of the pitcher's pitch types. Must be between -1 and 1. Negative values represent throwing each pitch type less often. Defaults to 0.05, a 5 percentage point increase.
                nonconstant_pt_increase_list (list(float)): The increase or decrease to each of the pitcher's pitch types. A different value can be specified for each pitch type. This must be a list of length equal to the number of pitch types that the pitcher threw that season with each list element between -1 and 1.
                pt_target_mix_list (list(float)): The target percentage thrown of each of a pitcher's pitch types. A different value can be specified for each pitch type. This must be a list of length equal to the number of pitch types that the pitcher threw that season with each list element between 0 and 1.
        
        Returns:
                pitcher_df (DataFrame): A DataFrame containing the percentage thrown and the run value per 100 pitches of each of a pitcher's pitch types that season, and the predicted difference in run value per 100 pitches based on the specified percentage thrown.
                None: Returned if there was an incorrectly provided parameter.
    '''
    # if the calculation is not recognized
    # return None
    if (calculation_type not in ['constant_increase', 'nonconstant_pt_increase', 'pt_equal_mix', 'pt_target_mix']):
        print(f"ERROR - INCOMPATIBLE PARAMETER: {calculation_type} IS NOT A RECOGNIZED CALCULATION.")
        print("Please set calculation_type to one of constant_increase, nonconstant_pt_increase, pt_equal, pt_target_percentage.\n")
        print("constant_increase: Every one of the pitcher's pitch types is thrown constant_increase_val percentage points more often.") 
        print("The constant_increase_val parameter must be a number between -1 and 1. Negative values represent throwing each pitch type less often.")
        print("This is the default calculation with a 5 percentage point increase the default value.\n")
        print("nonconstant_pt_increase: The use of each pitch type in a pitcher's mix is increased or decreased by a distinct percentage point amount specified in nonconstant_pt_increase_list.")
        print("The nonconstant_pt_increase_list parameter must be a list of numbers between -1 and 1 with length equal to the number of pitch types that the pitcher threw that season.\n")
        print("pt_equal_mix: Each of a pitcher's pitch types is thrown with equal proportion.")
        print("The proportion that each pitch type will be thrown is 1/n, where n is the number of pitch types that the pitcher threw that season.\n")
        print("pt_target_mix: Each of a pitcher's pitch types is thrown a distinct proportion of the time specified in pt_target_mix_list.")
        print("The pt_target_mix_list parameter must be a list of numbers between 0 and 1 with length equal to the number of pitch types that the pitcher threw that season.")
        print("A warning message will appear if the proportions don't sum to 1.\n")
        return None
    
    # if the season is not recognized
    # return None
    if (season not in [2021, 2022, 2023]):
        print(f"ERROR - INCOMPATIBLE PARAMETER: {season} IS NOT A RECOGNIZED SEASON.")
        print("Please set season to one of 2021, 2022, 2023, provided as a number.")
        return None
    
    # if the pitcher is not recognized
    # return None
    if (pitcher_id not in list(pitcher_full_seasons.index.unique(level='pitcher'))):
        print(f"ERROR - INCOMPATIBLE PARAMETER: {pitcher_id} IS NOT A RECOGNIZED PITCHER.")
        print("Please set pitcher_id to the MLBAM ID of a pitcher who appeared between 2021-2023, provided as a number.") 
        print("The function use_pybaseball_playerid_lookup will let you search for MLBAM ID by the pitcher's name.")
        return None
    
    # filter the data to include only the pitcher's pitch types that season
    pitcher_df = pitcher_full_seasons.loc[(season, slice(None), pitcher_id, slice(None)), :]
    
    # if all pitch types thrown equally
    if (calculation_type == 'pt_equal_mix'):
        # store array of percenatge thrown of each of the pitcher's pitch types
        delta_percentage_thrown_array = pitcher_df['pct'].to_numpy()
        # subtract actual percentage thrown from 1/n where n the number of pitch types the pitcher threw
        # get amount of change in percentage points needed to throw the pitch type 1/n
        delta_percentage_thrown_array = (1/len(pitcher_df.index)) - delta_percentage_thrown_array
    # if each pitch type will have different amount of change or different target proportion
    elif ((calculation_type == 'nonconstant_pt_increase') | (calculation_type == 'pt_target_mix')):
        # if each pitch type thrown a different target proportion
        if (calculation_type == 'pt_target_mix'):
            # if no parameter specified
            if (pt_target_mix_list == None):
                print("ERROR - REQUIRED PARAMETER NOT PROVIDED: Please provide pt_target_mix_list as a list of numbers between 0 and 1.")
                return None
            # if the parameter specified is not a list
            if (type(pt_target_mix_list) != list):
                print("ERROR - INCOMPATIBLE PARAMETER: pt_target_mix_list must be a list.")
                return None
            # if any element in the list not a number
            for percentage in pt_target_mix_list:
                if ((type(percentage) != float) & (type(percentage) != int)):
                    print("ERROR - INCOMPATIBLE PARAMETER: pt_target_mix_list must be a list of numbers.")
                    return None
            # if the list is not the same length as the number of pitch types the pitcher threw
            if (len(pt_target_mix_list) != len(pitcher_df.index)):
                print("ERROR - INCOMPATIBLE PARAMETER: pt_target_percentage_list is not of length equal to the number of pitch types that the pitcher threw that season.")
                print("The function view_pitch_types will show you the pitch types that the pitcher threw that season.")
                return None
            # if any target porportion not between 0 and 1
            if ((min(pt_target_mix_list) < 0) | (max(pt_target_mix_list) > 1)):
                print("ERROR - INCOMPATIBLE PARAMETER: Target percentages in pt_target_percentage_list must be between 0 and 1.")
                return None
            # if the target proportions do not sum to 1
            if (abs(sum(pt_target_mix_list) - 1) > 0.001):
                # print a warning
                # return calculation normally
                print("WARNING - TARGET PROPORTIONS DO NOT SUM TO 1.")
            # store array of percenatge thrown of each of the pitcher's pitch types
            delta_percentage_thrown_array = pitcher_df['pct'].to_numpy()
            # subtract actual percentage thrown from specified proportion
            # get amount of change in percentage points needed to throw the pitch type target proportion
            delta_percentage_thrown_array = np.array(pt_target_mix_list) - delta_percentage_thrown_array
        # else each pitch type thrown a different increase or decrease
        else:
            # if no parameter specified
            if (nonconstant_pt_increase_list == None):
                print("ERROR - REQUIRED PARAMETER NOT PROVIDED: Please provide nonconstant_pt_increase_list as a list of numbers between -1 and 1.")
                return None
            # if the parameter specified is not a list
            if (type(nonconstant_pt_increase_list) != list):
                print("ERROR - INCOMPATIBLE PARAMETER: pt_target_percentage_list must be a list.")
                return None
            # if any element in the list not a number
            for increase in nonconstant_pt_increase_list:
                if ((type(increase) != float) & (type(increase) != int)):
                    print("ERROR - INCOMPATIBLE PARAMETER: nonconstant_pt_increase_list must be a list of numbers.")
                    return None
            # if the list is not the same length as the number of pitch types the pitcher threw
            if (len(nonconstant_pt_increase_list) != len(pitcher_df.index)):
                print("ERROR - INCOMPATIBLE PARAMETER: nonconstant_pt_increase_list is not of length equal to the number of pitch types that the pitcher threw that season.")
                print("The function view_pitch_types will show you the pitch types that the pitcher threw that season.")
                return None
            # if any increase not between -1 and 1
            if ((min(nonconstant_pt_increase_list) < -1) | (max(nonconstant_pt_increase_list) > 1)):
                print("ERROR - INCOMPATIBLE PARAMETER: Increases or decreases in nonconstant_pt_increase_list must be between -1 and 1.")
                return None
            # amount of change is in provided list
            # store directly as array
            delta_percentage_thrown_array = np.array(nonconstant_pt_increase_list)
    # else the calculation is an increase applied to all pitch types
    else:
        # if the provided increase is not a number
        if ((type(constant_increase_val) != float) & (type(constant_increase_val) != int)):
            print("ERROR - INCOMPATIBLE PARAMETER: constant_increase_val must be a number.")
            return None
        # if the provided increase is not between -1 and 1
        if ((constant_increase_val < -1) | (constant_increase_val > 1)):
            print("ERROR - INCOMPATIBLE PARAMETER: Increase or decrease specified in constant_increase_val must be between -1 and 1.")
            return None
        # amount of change is provided number
        # create array with provided increase for each pitch type the pitcher threw
        delta_percentage_thrown_array = np.repeat(constant_increase_val, len(pitcher_df.index))
    
    # create list of columns except for the pitch usage
    # pitch usage is determined above
    columns = ['is_CU', 'is_FC', 'is_FF', 'is_FS', 'is_KC', 'is_SI', 'is_SL', 'is_ST', 'is_SV',
           'is_second', 'is_third', 'is_fourth', 'is_fifth_to_eighth',
          'std_pitch_velocity', 'std_spin_rate', 'std_spin_angle_adj', 'std_extension',
           'std_h_release_point_adj', 'std_v_release_point', 'std_h_movement_adj', 'std_v_movement']
    
    # convert columns to array
    pitcher_array = pitcher_df[columns].to_numpy().reshape(len(pitcher_df.index), len(columns))
    # standardize difference in percentage thrown
    delta_percentage_thrown_array = (delta_percentage_thrown_array - model_samples['pct_diff'].mean()) / model_samples['pct_diff'].std()
    # reshape difference in percentage thrown to add to columns
    delta_percentage_thrown_array = delta_percentage_thrown_array.reshape(-1, 1)
    # merge difference in percentage thrown with other columns
    pitcher_array = np.concatenate((delta_percentage_thrown_array, pitcher_array), axis=1)
    # get interaction terms between each variable
    pitcher_array_ = PolynomialFeatures(interaction_only=True, include_bias=False).fit_transform(pitcher_array)
    
    # predict the change to run value per 100 pitches
    pitcher_df['pred_RV_per_100_diff'] = model.predict(pitcher_array_)
    
    # filter the data to keep (actual) percentage thrown, run value per 100 pitches and predicted difference to run value per 100 pitches
    pitcher_df = pitcher_df[['pct', 'RV_per_100', 'pred_RV_per_100_diff']]
    # return the dataframe
    return(pitcher_df)
```

# Cleaning the Data

```python
# remove pitches without a change in RE288
pitches = pitches[pd.isna(pitches['delta_run_exp']) == False]
# remove pitches with more than 3 balls or more than 2 strikes
pitches = pitches[(pitches['balls'] <= 3) & (pitches['strikes'] <= 2)]
# number of pitches in the dataset
len(pitches.index)
```

```python
# number of pitches of each pitch type in dataset
pitches.groupby(['pitch_type', 'pitch_name']).size().sort_values(ascending=False)
```

```python
# replace forkball with split-finger
pitches = pitches.replace('FO', 'FS')
pitches = pitches.replace('Forkball', 'Split-Finger')
```

```python
# filter to keep only selected pitch types
pitches = pitches[pitches['pitch_type'].isin(['CH', 'CU', 'FC', 'FF', 'FS', 'KC', 'SI', 'SL', 'ST', 'SV'])]
# number of pitches in dataset
len(pitches.index)
```

```python
# add a binary column indicating the pitch type
# for calculating percentage thrown
for pitch_type in ['CH', 'CU', 'FC', 'FF', 'FS', 'KC', 'SI', 'SL', 'ST', 'SV']:
    # if the pitch type is equal to the desired pitch type
    # set the pitch type column to true
    pitches[pitch_type] = pitches['pitch_type'] == pitch_type
    # convert the column from boolean to a binary column
    pitches[pitch_type] = pitches[pitch_type].astype(int)
```

# Creating Sample Numbers

```python
# sort dataset by date of game, game id, plate appearance of game, then the pitch number of the plate appearance
pitches = pitches.sort_values(by=['game_date', 'game_pk', 'at_bat_number', 'pitch_number'], ignore_index=True)

pitches.head()
```

```python
# create column containing season pitch count
# .cumcount() starts at 0, add 1 to start count at 1
pitches['season_pitch_number'] = pitches.groupby(['game_year', 'player_name', 'pitcher']).cumcount() + 1
```

```python
# create DataFrame holding number of pitches thrown in season by pitcher
# number of pitches thrown is equivalent to the maximum of the season pitch count
pitcher_seasons = pitches.groupby(['game_year', 'player_name', 'pitcher']).agg({'season_pitch_number': 'max'})

# view the first 10 pitchers' season pitch counts
pitcher_seasons.head(10)
```

```python
# set the sample size
sample_size = 1000

# add column for the number of samples of the sample size for each player
# use // for integer division, round to lowest whole number
pitcher_seasons['num_samples'] = pitcher_seasons['season_pitch_number'] // sample_size

# subset columns to keep only game_year, player_name, pitcher, num_samples
# game_year, player_name, pitcher are on the index and are kept by default
pitcher_seasons = pitcher_seasons[['num_samples']]

pitcher_seasons.head()
```

```python
# add column containing number of samples of sample size to pitches DataFrame
# use inner join as all pitchers have a number of samples
pitches = pitches.join(pitcher_seasons, on=['game_year', 'player_name', 'pitcher'], how='inner')

# view DataFrame with number of samples
pitches[['game_date', 'game_year', 'player_name', 'batter', 'pitcher', 'home_team', 'game_pk', 'season_pitch_number', 'num_samples']].head()
```

```python
# create a DataFrame filtered only for pitchers who have at least one sample of 1000 pitches
# this removes pitchers who have thrown less than 1000 pitches
pitches_min_1000 = pitches[pitches['num_samples'] > 0]
# filter to have only pitches in a sample
pitches_min_1000 = pitches_min_1000[pitches_min_1000['season_pitch_number'] <= (pitches_min_1000['num_samples'] * sample_size)]

# view number of pitches that were thrown in filtered dataset
pitches_min_1000.groupby(['game_year', 'player_name', 'pitcher']).agg({'season_pitch_number': 'max'}).head()
```

```python
# view max number of pitches that were thrown in filtered dataset
# sort highest-to-lowest by filtered pitch count, then alphabetically by name
pitches_min_1000.groupby(['game_year', 'player_name', 'pitcher']).agg({'season_pitch_number': 'max'}).sort_values(by=['season_pitch_number', 'player_name', 'pitcher'], ascending=[False, True, True]).head()
```

```python
# cast num_samples to int
# must be int to group
pitches_min_1000['num_samples'] = pitches_min_1000['num_samples'].apply(np.int64)

# create sample number for each pitch
# assign all pitches to sample 1 first
pitches_min_1000['sample_number'] = 1
# if season_pitch_number is greater than 1000 and less than 1000 then assign to sample 2
# this is there are no pitchers with more than three samples of 1000 pitches
pitches_min_1000.loc[(pitches_min_1000['season_pitch_number'] > 1000) & (pitches_min_1000['season_pitch_number'] <= 2000), 'sample_number'] = 2
# if season_pitch_number is greater than 1000 then assign to sample 2
# there will be no pitches above 3000
# this is there are no pitchers with more than three samples of 1000 pitches
pitches_min_1000.loc[(pitches_min_1000['season_pitch_number'] > 2000), 'sample_number'] = 3

# view first 10 pitches with sample number
pitches_min_1000[['game_date', 'game_year', 'player_name', 'batter', 'pitcher', 'home_team', 'game_pk', 'season_pitch_number', 'num_samples', 'sample_number']].head(10)
```

# Pitch Metrics

```python
# create DataFrame average of Statcast metrics for each pitch type in the season
# include pitcher handedness in index
pitch_type_metrics = pitches.groupby(['game_year', 'player_name', 'pitcher', 'pitch_type', 'p_throws']).agg(pitch_velocity=('release_speed', 'mean'),
                                                                                                            spin_rate=('release_spin_rate', 'mean'),
                                                                                                            spin_angle=('spin_axis', 'mean'),
                                                                                                            extension=('release_extension', 'mean'),
                                                                                                            h_release_point=('release_pos_x', 'mean'),
                                                                                                            v_release_point=('release_pos_z', 'mean'),
                                                                                                            h_movement=('pfx_x', 'mean'),
                                                                                                            v_movement=('pfx_z', 'mean'))
# reset the index
pitch_type_metrics = pitch_type_metrics.reset_index()
# set the index to be all previous index columns except for the pitcher handedness
pitch_type_metrics = pitch_type_metrics.set_index(['game_year', 'player_name', 'pitcher', 'pitch_type'])
pitch_type_metrics.head()
```

```python
# create column to have spin angle be adjusted for handedness
# set the adjusted spin angle to be the original spin angle
pitch_type_metrics['spin_angle_adj'] = pitch_type_metrics['spin_angle']
# if the pitcher is left-handed then subtract the spin angle from 360
# this sets all spin angles to be as if the pitcher were right-handed
pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'spin_angle_adj'] = 360 - pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'spin_angle']

# create column to have horizontal release point be adjusted for handedness
# set the horizontal release point to be the original horizontal release point
pitch_type_metrics['h_release_point_adj'] = pitch_type_metrics['h_release_point']
# if the pitcher is left-handed then multiply the horizontal release point by -1
# this sets all spin angles to be as if the pitcher were right-handed
pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'h_release_point_adj'] = (-1) * pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'h_release_point']

# create column to have hroizontal movement be adjusted for handedness
# set the adjusted horizontal movement to be the original horizontal movement
pitch_type_metrics['h_movement_adj'] = pitch_type_metrics['h_movement']
# if the pitcher is left-handed then multiply the horizontal movement by -1
# this sets all spin angles to be as if the pitcher were right-handed
pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'h_movement_adj'] = (-1) * pitch_type_metrics.loc[pitch_type_metrics['p_throws'] == 'L', 'h_movement']

pitch_type_metrics.head()
```

```python
# check that the adjusted columns are properly scaled
# the values for each pitch type in the '_adj' columns in each pitch type should be similar
pitch_type_metrics.groupby(['pitch_type', 'p_throws']).agg(spin_angle=('spin_angle', 'mean'),
                                                          h_release_point=('h_release_point', 'mean'),
                                                          h_movement=('h_movement', 'mean'),
                                                          spin_angle_adj=('spin_angle_adj', 'mean'),
                                                          h_release_point_adj=('h_release_point_adj', 'mean'),
                                                          h_movement_adj=('h_movement_adj', 'mean'))
```

# Fitting the Model

```python
# create DataFrame for fitting the model
# remove pitch types that were thrown less than eight percent of the time in the full season
model_samples = samples[samples['pct_full_season'] >= 0.08]
```

```python
# standardize the variables for fitting LASSO regression
# subtract the mean of the column, then divide by the standard deviation
model_samples['std_RV_per_100_diff'] = (model_samples['RV_per_100_diff'] - model_samples['RV_per_100_diff'].mean())/model_samples['RV_per_100_diff'].std()
model_samples['std_pct_diff'] = (model_samples['pct_diff'] - model_samples['pct_diff'].mean())/model_samples['pct_diff'].std()

model_samples['std_pitch_velocity'] = (model_samples['pitch_velocity'] - model_samples['pitch_velocity'].mean())/model_samples['pitch_velocity'].std()
model_samples['std_spin_rate'] = (model_samples['spin_rate'] - model_samples['spin_rate'].mean())/model_samples['spin_rate'].std()
model_samples['std_spin_angle_adj'] = (model_samples['spin_angle_adj'] - model_samples['spin_angle_adj'].mean())/model_samples['spin_angle_adj'].std()
model_samples['std_extension'] = (model_samples['extension'] - model_samples['extension'].mean())/model_samples['extension'].std()
model_samples['std_h_release_point_adj'] = (model_samples['h_release_point_adj'] - model_samples['h_release_point_adj'].mean())/model_samples['h_release_point_adj'].std()
model_samples['std_v_release_point'] = (model_samples['v_release_point'] - model_samples['v_release_point'].mean())/model_samples['v_release_point'].std()
model_samples['std_h_movement_adj'] = (model_samples['h_movement_adj'] - model_samples['h_movement_adj'].mean())/model_samples['h_movement_adj'].std()
model_samples['std_v_movement'] = (model_samples['v_movement'] - model_samples['v_movement'].mean())/model_samples['v_movement'].std()
```

```python
# view the descriptive statistics
model_samples.describe()
```

```python
# intialize a list of the variables the model will be fit on
model_columns = ['std_pct_diff', 
           'is_CU', 'is_FC', 'is_FF', 'is_FS', 'is_KC', 'is_SI', 'is_SL', 'is_ST', 'is_SV',
           'is_second', 'is_third', 'is_fourth', 'is_fifth_to_eighth',
          'std_pitch_velocity', 'std_spin_rate', 'std_spin_angle_adj', 'std_extension',
           'std_h_release_point_adj', 'std_v_release_point', 'std_h_movement_adj', 'std_v_movement']

# set the testing split
test_split = 0.2
# set the number of cross-validation folds
cv_folds = 5
```

```python
# create an array with the response variable
y = model_samples['RV_per_100_diff'].to_numpy()  
# create an array with the explanatory variables
# reshape to make a row in the dataset a row in the array
x = model_samples[model_columns].to_numpy().reshape(len(model_samples), len(model_columns))
# create an array with second-degree interaction terms between each of the main effect variables
x_ = PolynomialFeatures(interaction_only=True, include_bias=False).fit_transform(x)

# split into test and training set with testing split
X_train, X_test, y_train, y_test = train_test_split(x_, y, test_size=test_split, random_state=42)

# fit a LASSO regression model using cross-validation to select the hyperparameter
# the number of folds is the number of cross-validation folds
model = LassoCV(cv=cv_folds, random_state=42).fit(X_train, y_train)
```

```python
# calculate the model accuracy on the training set
# predict on the observations in the training set
y_pred_train = model.predict(X_train)

# calculate the residual between the predicted and observed values
residuals_train = np.subtract(y_pred_train, y_train)
# square the residuals
# then take the mean to get MSE
mse_train = np.square(residuals_train).mean()
# take the square root of the mean to get RMSE
rmse_train = np.sqrt(mse_train)
# view the training RMSE
print(f"Training RMSE: {rmse_train}")

# view the training R^2
print(f"Training R2: {r2_score(y_train, y_pred_train)}")
```

```python
# calulate the model accuracy on the testing set
# predict on the obervations in the testing set
y_pred_test = model.predict(X_test)

# calculate the residual between the predicted and observed values
residuals_test = np.subtract(y_pred_test, y_test)
# square the residuals
# then take the mean to get MSE
mse_test = np.square(residuals_test).mean()
# take the square root of the mean to get RMSE
rmse_test = np.sqrt(mse_test)
# view the testing RMSE
print(f"Test RMSE: {rmse_test}")

# view the testing R^2
print(f"Test R2: {r2_score(y_test, y_pred_test)}")
```

```python
# create a list of variable names in the model
# including the interaction terms
# copy the main effect variables to a new list
model_variables = model_columns.copy()
# for each variable in the list of main effect variables
for i in range(len(model_columns)):
    # for each variable in the list of main effect variables after that one 
    for j in range((i+1),len(model_columns)):
        # append to the list the interaction term between variables
        model_variables.append(model_columns[i]+'*'+model_columns[j])
```

```python
# create list of variable names without prefix
# copy variable names with prefix
variable_names = model_variables.copy()
# for each model variable name
for i in range(len(model_variables)):
    # remove 'std_'
    variable_names[i] = re.sub('std_', '', model_variables[i])
```

```python
# view the intercept of the model
print(f"intercept: {model.intercept_}")
# for each variable
for i in range(len(model.coef_)):
    # if it is nonzero
    if model.coef_[i] != 0:
        # view the coefficient of the variable
        print(f"{model_variables[i]}: {model.coef_[i]}")
```

```python
# create a DataFrame of the nonzero coefficients of the model
coefficients_df = pd.DataFrame(data={'model_variable': [model_variables[i] for i in range(len(model.coef_)) if model.coef_[i] != 0],
                                  'coefficient': [model.coef_[i] for i in range(len(model.coef_)) if model.coef_[i] != 0],
                                    'variable': [variable_names[i] for i in range(len(model.coef_)) if model.coef_[i] != 0]})
# add the model intercept
coefficients_df = coefficients_df.append({'model_variable': 'intercept', 'coefficient': model.intercept_, 'variable': 'intercept'}, ignore_index=True)
# sort largest-to-smallest in coefficient
coefficients_df = coefficients_df.sort_values(by='coefficient', ascending=False, ignore_index=True)

coefficients_df
```

```python
# add unscaled coefficients of main effect variables to DataFrame
# divide by standard deviation of the variable
# set each unscaled coefficient to be NaN
coefficients_df['unscld_coefficient'] = np.nan
# unscale percentage difference
# multiply by 0.01 to interpret on unit scale
coefficients_df.loc[coefficients_df['model_variable'] == 'std_pct_diff', 'unscld_coefficient'] = (coefficients_df.loc[coefficients_df['model_variable'] == 'std_pct_diff', 'coefficient'] / model_samples['pct_diff'].std()) * 0.01
# unscale h_release_point_adj
coefficients_df.loc[coefficients_df['model_variable'] == 'std_h_release_point_adj', 'unscld_coefficient'] = coefficients_df.loc[coefficients_df['model_variable'] == 'std_h_release_point_adj', 'coefficient'] / model_samples['h_release_point_adj'].std()
# unscale h_movement_adj
coefficients_df.loc[coefficients_df['model_variable'] == 'std_h_movement_adj', 'unscld_coefficient'] = coefficients_df.loc[coefficients_df['model_variable'] == 'std_h_movement_adj', 'coefficient'] / model_samples['h_movement_adj'].std()
# copy intercept
coefficients_df.loc[coefficients_df['model_variable'] == 'intercept', 'unscld_coefficient'] = coefficients_df.loc[coefficients_df['model_variable'] == 'intercept', 'coefficient']
# view all coefficients in DataFrame
coefficients_df
```

```python
# merge the pitch type metrics to the pitcher_full_season DataFrame
pitcher_full_seasons = pitcher_full_seasons.join(pitch_type_metrics, how='inner')
# remove pitch types with NaN spin rate or spin angle
pitcher_full_seasons = pitcher_full_seasons[(pd.isna(pitcher_full_seasons['spin_rate']) == False) & (pd.isna(pitcher_full_seasons['spin_angle_adj']) == False)]
```

```python
# standardize the variables for fitting LASSO regression
pitcher_full_seasons['std_pitch_velocity'] = (pitcher_full_seasons['pitch_velocity'] - model_samples['pitch_velocity'].mean())/model_samples['pitch_velocity'].std()
pitcher_full_seasons['std_spin_rate'] = (pitcher_full_seasons['spin_rate'] - model_samples['spin_rate'].mean())/model_samples['spin_rate'].std()
pitcher_full_seasons['std_spin_angle_adj'] = (pitcher_full_seasons['spin_angle_adj'] - model_samples['spin_angle_adj'].mean())/model_samples['spin_angle_adj'].std()
pitcher_full_seasons['std_extension'] = (pitcher_full_seasons['extension'] - model_samples['extension'].mean())/model_samples['extension'].std()
pitcher_full_seasons['std_h_release_point_adj'] = (pitcher_full_seasons['h_release_point_adj'] - model_samples['h_release_point_adj'].mean())/model_samples['h_release_point_adj'].std()
pitcher_full_seasons['std_v_release_point'] = (pitcher_full_seasons['v_release_point'] - model_samples['v_release_point'].mean())/model_samples['v_release_point'].std()
pitcher_full_seasons['std_h_movement_adj'] = (pitcher_full_seasons['h_movement_adj'] - model_samples['h_movement_adj'].mean())/model_samples['h_movement_adj'].std()
pitcher_full_seasons['std_v_movement'] = (pitcher_full_seasons['v_movement'] - model_samples['v_movement'].mean())/model_samples['v_movement'].std()
```

# Example 1 - Average Expected Change by Pitch Type

```python
# create list of columns except for the pitch usage
# pitch usage will be set depending on the desired prediction
example_columns = ['is_CU', 'is_FC', 'is_FF', 'is_FS', 'is_KC', 'is_SI', 'is_SL', 'is_ST', 'is_SV',
           'is_second', 'is_third', 'is_fourth', 'is_fifth_to_eighth',
          'std_pitch_velocity', 'std_spin_rate', 'std_spin_angle_adj', 'std_extension',
           'std_h_release_point_adj', 'std_v_release_point', 'std_h_movement_adj', 'std_v_movement']
# convert example columns to array for prediction
example_array = pitcher_full_seasons[example_columns].to_numpy().reshape(len(pitcher_full_seasons), len(example_columns))
```

# Example 1a - 5 Percentage Point Increase

```python
# create array of increased usage by 5 percentage points
incr_by_five_pct_array = np.repeat(0.05, len(pitcher_full_seasons.index))
# standardize 5 percentage point increase
# use model_samples mean and standard deviation
# this the values that the model was fit on
incr_by_five_pct_array = (incr_by_five_pct_array - model_samples['pct_diff'].mean()) / model_samples['pct_diff'].std()
# reshape for adding to the array of other columns
incr_by_five_pct_array = incr_by_five_pct_array.reshape(-1, 1)
# create prediction array 
example_incr_by_five_pct_array = np.concatenate((incr_by_five_pct_array, example_array), axis=1)
# get interaction terms
example_incr_by_five_pct_array_ = PolynomialFeatures(interaction_only=True, include_bias=False).fit_transform(example_incr_by_five_pct_array)
    
# predict change in RV_per_100 when using the pitch type 5 percentage points more often
pitcher_full_seasons['pred_RV_per_100_diff_incr_five_pct'] = model.predict(example_incr_by_five_pct_array_)
```

```python
# get average expected change by pitch type using the pitch type 5 percentage points more often
pitcher_full_seasons.groupby('pitch_type').agg(pred_RV_per_100_increase_five_pct=('pred_RV_per_100_diff_incr_five_pct', 'mean')).sort_values(by='pred_RV_per_100_increase_five_pct', ascending=False)
```
